# GroupBy Hint

```python
from pathlib import Path  # Python Standard Library
import pandas as pd  # pip install pandas openpyxl
```

```python
# Read Excel file (Japan_2021.xlsx) into a pandas DataFrame.
this_dir = Path(__file__).parent if "__file__" in locals() else Path.cwd()
wb_file_path = this_dir / "Japan_2021.xlsx"
df = pd.read_excel(
    io=wb_file_path,
    engine="openpyxl",
    sheet_name="Overview",
    skiprows=3,
    usecols="B:L",
    nrows=105,
)
```

```python
df.head()
```

```python
df["Sales"].sum()
```

```python
by_segment = df.groupby(by=["Segment"])[["Sales"]].sum()
by_segment.sum()
```

```python
by_segment_and_product = df.groupby(by=["Segment", "Product"], dropna=False)[["Sales"]].sum()
by_segment_and_product
```
